```python
from collections import defaultdict
#from operator import itemgetter
```

```python
d9id_ftid = defaultdict(set)
with open("2016_07_16_clk.txt", 'r') as fh:
    for line in fh:
        v = line.strip().split("|")
        ftid = v[0]
        d9id = v[1][3:]
        
        if d9id == "":
            continue
        d9id_ftid[d9id].add(ftid)        
```

```python
# sort by length of value, print key only
sorted(d9id_ftid, key = lambda i: len(d9id_ftid[i]), reverse = True)
```

```python
# sort by length of value, print both key and value
for d in sorted(d9id_ftid, key = lambda i: len(d9id_ftid[i]), reverse = True):
    print "{}|{}".format(d, ",".join(d9id_ftid[d]))
```
